```python
# read the data into a pandas dataframe
dfagg = df

# convert Date Rptd column to datetime format and extract Month and Year columns
dfagg['Date Rptd'] = pd.to_datetime(dfagg['Date Rptd'])
dfagg['Month'] = dfagg['Date Rptd'].dt.month
dfagg['Year'] = dfagg['Date Rptd'].dt.year

# Group by Month, Year, AREA, AREA NAME, Vict Sex, Crime cd, Crime cd desc, ZipCode,
# population, POP_SQMI, SQMI and get the counts of crimes
counts = dfagg.groupby(['Month',
                     'Year',
                     'AREA',
                     'AREA NAME',
                     'Vict Sex',
                     'CATEGORY',
                     'ZIP_CODE',
                     'POPULATION',
                     'POP_SQMI',
                     'SQMI']).agg({'Date Rptd': 'count'})

# Rename the Date Rptd column to Crime Count
counts = counts.rename(columns={'Date Rptd': 'Crime Count'})

# Reset the index to get the columns back
counts = counts.reset_index()

counts.head()
```

```python
counts.to_csv('Crime_Data_2020_to_Present_AGG.csv', index=False)
```

```python
parksDF = pd.read_csv('Clean_Data_To_Join/PARK_FINAL_CLEANED.csv')
# pivot the dataframe to get park types as columns
parksDF = parksDF.pivot_table(index='Zip', columns='Park Type', values='Count', fill_value=0)

# reset the index to get Zip as a column
parksDF = parksDF.reset_index()
parksDF = parksDF.rename(columns={'Zip': 'ZIP_CODE'})
parksDF.head()
```

```python
crimeDF = counts.merge(parksDF, on='ZIP_CODE', how='left').fillna(0)
crimeDF.head()
```

```python
schoolsDF = pd.read_csv('Clean_Data_To_Join/LA_SCHOOLS_CLEANED_FINAL.csv')
# pivot the dataframe to get park types as columns
schoolsDF = schoolsDF.pivot_table(index='Zip Code', columns='School type', values='COUNT of Zip Code', fill_value=0)

# reset the index to get Zip as a column
schoolsDF = schoolsDF.reset_index()
schoolsDF = schoolsDF.rename(columns={'Zip Code': 'ZIP_CODE'})
schoolsDF.head()
```

```python
crimeDF = crimeDF.merge(schoolsDF, on='ZIP_CODE', how='left').fillna(0)
crimeDF.head()
```

```python
from datetime import datetime

activeBusDF = pd.read_csv('Clean_Data_To_Join/Listing_of_Active_Businesses_Clean_Counts.csv')
activeBusDF.drop('NAICS_2', axis=1, inplace=True)
# convert 'Year' column to datetime and format as YYYY
activeBusDF['Year'] = pd.to_datetime(activeBusDF['Year'], format='%Y').dt.strftime('%Y')
activeBusDF['Month'] = pd.to_datetime(activeBusDF['Month'], format='%m').dt.strftime('%m')
# filter for Year 2020 and above
activeBusDF = activeBusDF[activeBusDF['Year'].astype(int) >= 2020]

# pivot the dataframe to get park types as columns
#activeBusDF = activeBusDF.pivot_table(index='ZIPCODE', columns='NAICS_2', values='Count', fill_value=0)
activeBusDF = activeBusDF.pivot_table(index=['ZIPCODE', 'Month', 'Year'], columns='Industry', values='Count', aggfunc='sum', fill_value=0)

# reset the index to get Zip as a column
activeBusDF = activeBusDF.reset_index()
activeBusDF = activeBusDF.rename(columns={'ZIPCODE': 'ZIP_CODE'})

activeBusDF.head()
```

```python
crimeDF['Year'] = pd.to_datetime(crimeDF['Year'], format='%Y').dt.strftime('%Y')
crimeDF['Month'] = pd.to_datetime(crimeDF['Month'], format='%m').dt.strftime('%m')
crimeDF.head()
```

```python
crimeDF_Bus = pd.merge(crimeDF, activeBusDF, on=['ZIP_CODE', 'Month', 'Year'], how='left').fillna(0)
crimeDF_Bus.head()
```

```python
PopulationWithGenderDF = pd.read_csv('Clean_Data_To_Join/Population_With_Gender.csv')
PopulationWithGenderDF = PopulationWithGenderDF.rename(columns={'Zip Code': 'ZIP_CODE'})
crimeDF_Bus_Pop = pd.merge(crimeDF_Bus, PopulationWithGenderDF, on=['ZIP_CODE'], how='left').fillna(0)
crimeDF_Bus_Pop.head()
```

```python
medianIncomeDF = pd.read_csv('Clean_Data_To_Join/Median_Household_Income_With_ZipCode.csv')
medianIncomeDF = medianIncomeDF.rename(columns={'Zip Code': 'ZIP_CODE'})
medianIncomeDF.drop('Community', axis=1, inplace=True)
crimeDF_Bus_Pop_Inc = pd.merge(crimeDF_Bus_Pop, medianIncomeDF, on=['ZIP_CODE'], how='left').fillna(0)
crimeDF_Bus_Pop_Inc.head()
```

```python
nonProfitDF = pd.read_csv('Clean_Data_To_Join/Non_Profit_Locs.csv')
nonProfitDF = nonProfitDF.rename(columns={'ZIP CODE': 'ZIP_CODE'})

crimeDF_Bus_Pop_Inc_nonprofit = pd.merge(crimeDF_Bus_Pop_Inc, nonProfitDF, on=['ZIP_CODE'], how='left').fillna(0)
crimeDF_Bus_Pop_Inc_nonprofit.head()

```

```python
crimeDF_Bus_Pop_Inc.to_csv('Crime_Data_2020_to_Present_FINAL.csv', index=False)
```
